# **Dataset analysis**

The dataset that will be used in this paper is the "NASA: Asteroid Classification" dataset created by CNEOS (Center For Near Earth Object Studies). This center calculated the orbits of asteroids and comets and the chances of their impact on the Earth. The dataset contains 23 columns. Each of them represented an independent variable in the analysis.

**neo_reference_id**

It represents the identification number given to asteroids. This variable will not be used in the analysis because it does not contain relevant data about the nature of the asteroid itself.

**name**

Indicates the name of the asteroid. This data will not be used in the analysis because it does not contain relevant data about the nature of the asteroid itself.

**absolute_magnitude**


The magnitude an observer would measure if he or she was at a distance of one astronomical unit from the asteroid and the Sun at a phase angle of zero. This parameter is also called the H parameter.


**est_dia_in_km_min**

This variable contains the estimated smallest possible radius of the asteroid in km.


**est_dia_in_km_max**

This variable contains the estimated maximum possible radius of the asteroid in km.


**relative_velocity_km_per_sec**

The column contains the relative velocity values ​​of the asteroid in km/s.
miss_dist.(kilometers)
It represents the distance between the asteroid and the Earth at the moment of passing.


**orbit_id**


Represents the identification number given to the orbits. This data will not be used in the analysis because it does not contain relevant data about the nature of the asteroid itself.
orbit_uncertainty
The degree of orbital unpredictability is a parameter used to estimate the approach of an asteroid to the earth, which affects the degree of potential risk of an asteroid impact.[4]


**minimum_orbit_intersection**

It is measured in astronomical units. It represents the distance between the two closest points of the oscillating orbits of two celestial bodies. NASA uses this parameter to classify asteroids as hazardous if this parameter is 0.05 AU or less.[14]


**jupiter_tisserand_invariant**

The Tisserand parameter is a quantity that remains almost the same when a celestial object encounters a planet or interplanetary body and thus relates the properties of the body in the phase before the encounter with the properties after the encounter. As Jupiter is the most massive planet in the Solar System, it exerts the greatest dynamic influence on asteroids. So Tisserand's Jupiter parameter is most often used.[15]


**epoch_osculation**
A parameter indicating the asteroid osculation epoch.


**eccentricity**
How is the orbit of an asteroid in the shape of an ellipse. Its eccentricity represents the degree of deviation from the circle based on the distance of its foci. If the value of this parameter is 0, the orbit is circular. From 0 to 1 it is elliptical, with a value of 1 it is parabolic, while with values ​​over 1 it is a hyperbolic orbit.[16]


**semi_major_axis**

This variable contains values ​​equal to half the head of the axis of the elliptical orbit as well as the mean distance from the Sun.[18]

**inclination**

It indicates the angle between the orbital plane and the reference plane.[17]

**asc_node_longitude**

The angle from the origin of the longitude of the reference plane to the entry node of the orbit, which represents the point in its orbit where the body crosses the reference plane towards the north. The time of this crossing is also called the epoch of the elements. [17]


**orbital_period**

This variable consists of data on the orbital period of the asteroid.
perihelion_distance
The distance of the point on an asteroid's orbit that is closest to the Sun from the Sun.


**perihelion_time**

Duration of perihelion.

**aphelion_dist**

The distance from the Sun to the point on an asteroid's orbit that is farthest from the Sun.

**mean_anomaly**

The mean object is used to define the mean anomaly. This is an object of the same orbital period as the observed object passing through perihelion and moving at a constant speed in a circular orbit with a radius equal to the semi-axis of the observed object
The mean anomaly is the angle between the line connecting the center of the orbit and the mean object and the line connecting the center of the orbit to perihelion. [18]


**mean_motion**
Asteroid speed.

**hazardous**
This variable contains the binary values ​​0 and 1. Where 0 indicates harmless asteroids and 1 indicates dangerous asteroids.

## **Data analysis and processing**
The dataset contains data on asteroids of which 83.9% are harmless while 16.1% of asteroids are dangerous. This means that the dataset is unbalanced because it contains more data from one group. This makes it difficult to create an accurate model that will classify asteroids in relation to the given parameters and not just assign a value based on the majority. Therefore, the data must be balanced before training the model.

Then the data was split into Train and Test data using the sklearn.model_selection.train_test_split function. In order for the data to be randomly distributed in these data groups, the parameter shuffle = True was used. In this project, the data used to train the neural network makes up 70% of the total data, while the testing data makes up 30% of the total data. Therefore, data on 3280 asteroids will be used for training, while 1407 will be used for testing. The training data can be further balanced in three ways. By reducing the data from the group of harmless asteroids so that their number is equal to the number of dangerous asteroids. However, then the number of data used for testing would be over three times smaller than the initial one, which will make it difficult for the algorithm to create an accurate classification model. Another way is to repeat the data from a less represented group in order to increase their number, but this does not provide new information about the data needed to implement the appropriate model. So, in this paper, the third method was used, which includes the creation of synthetic data of dangerous asteroids. This was done using the SMOTE technique (Synthetic Minority Oversampling technique) within the imblearn library.[13]

### Selection of independent variables
#### Pearson correlation method
Pearson's correlation is used to determine the correlation between variables. It can have a value in the range from -1 to 1, where 0 means that there is no correlation, while 1 is a total positive correlation and -1 is a total negative correlation. A positive correlation indicates that the value of an arbitrary variable A will increase with an increase in variable B, while a negative correlation indicates that A decreases with an increase in the value of variable B.[3]

With the help of Pearson's correlation, we can determine the correlation between variables. This can be most simply represented with a graphical representation of the correlation matrix. From this matrix it can be seen that the correlation between the pairs of variables jupiter_tisserand_invariant and mean_motion, epoch_osculation and perihelion_time, semi_major_axis and aphelion_dist, and est_dia_in_km_min and est_dia_in_km_max is extremely strong and approximately one. Which means that one variable can be selected from each pair to avoid redundancy.

#### ANOVA and LASSO hybrid method
LASSO (Least Absolute Shrinkage and Selection Operator) is a statistical formula used for model regularization and variable selection. Based on this method, the variables orbit_uncertainty, absolute_magnitude, est_dia_in_km_max, relative_velocity_km_per_sec and minimum_orbit_intersection were selected.

### Gradient Boosting Classification method
The Gradient Boosting algorithm is similar to the Ada Boosting Decision Tree algorithm because both are a combination of Boosting and Decision Tree algorithms. What is specific about the Gradient Boosting algorithm is that a key part of the algorithm is finding the best hyperparameters that allow to build the best possible model for a given data set. However, finding the best hyperparameters takes a lot of time. One of the hyperparameters is the Learning rate which controls how fast the algorithm changes. Smaller values ​​reduce the possibility of overfitting, but the algorithm needs more time to execute, while the algorithm is faster with larger values, but it may not be able to successfully determine the rules based on which it classifies the data. Number of trees represents the optimal number of trees for the model. Subsampling indicates what percentage of the data from the training set is used to train a single tree. [25]
In this paper, the best hyperparameters for this data were determined using the RandomSearch technique.

###Naive Bayes method
Bayes' theorem is of particular significance in statistics and machine learning. Based on this theorem, a naive Bayesian classifier was formed, which is a simple but powerful machine learning algorithm.

In a data set with n instances each containing p attributes, each instance belongs to only one class y. Most predictive machine learning models calculate a number representing the degree to which an instance belongs to class xi and yj. In case of binary classification, where there is only positive and negative class y, a limit is set for the values ​​of this number, on the basis of which class y is determined. If that value is less than the limit, it will belong to the negative class y, that is, the value of the y parameter for this instance will be 0. If that value is greater than the limit, it will belong to the positive class y and the value of the y parameter for this instance will be 1.

A naive Bayesian classifier uses the probability of an instance belonging to a class when classifying objects. In the model of this classifier, it is assumed that all attributes are independent, which is a disadvantage of this algorithm when used with data sets where parameters influence each other. However, this algorithm is often used in machine learning and on many data sets it shows a high efficiency that can be compared to more advanced algorithms. It has practical applications in text classification, medical diagnostics, and system performance management. [5,6,7]

In this project, the MultinomialNB model from the sklearn library was used for the classification of asteroids with the Naive Bayesian algorithm. ​​Over the asteroid data set where there is a dependence between certain parameters, the success of this algorithm is 68.79% on the data from the test set, while the success on the training data is 72.81%. ### Logistic Regression method
Logistic regression is one of the most important statistical methods used in areas such as health analysis, medical statistics, ecology, econometrics, and statistics in the social sciences. It is similar to normal and linear regression, but as a result it gives binary values ​​(0 or 1). So it is used to solve binary classification problems. A logistic regression model predicts the probability that the y value of the parameters is 1 based on a set of attribute values.
Regression serves to determine the value of the dependent variable based on the value of the independent variables of one instance.

\begin{equation}  
\hat{y_i} = \beta_0+ \beta_1x_{i1} + \beta_2x_{i2} + ...  \beta_p*x_{ip}
\end{equation}  

Where $\hat{y}$ is the sum of all terms in the regression, where each term is the product of the coefficient $\beta_j,( j\in [0,p])$ and the value of the attribute $x_{ij}, ( i ∈[1,n] )$. Logistic regression uses the following relationship:


\begin{equation}  
ln(\frac{𝜇_i}{1-𝜇_i}) = \beta_0+ \beta_1x_{i1} + \beta_2x_{i2} + ...  \beta_p*x_{ip}
\end{equation}  

In which 𝜇 indicates the probability that the value of the dependent variable y is equal to 1. So, for binary classification using logistic regression, it is necessary to determine how to distribute the data into two groups. [8]

In this paper, the Logistic Regression model from the sklearn library was used to classify asteroids using the Logistic Regression algorithm. For this data set with the attributes minimum_orbit_intersection, est_dia_in_km_max, absolute_magnitude, perihelion_distance, relative_velocity_km_per_sec, and orbit_uncertainty which we can respectively denote by $x1, x2, x3, ... x5, x6$ . The resulting relationship used in logistic regression on this data set is:


\begin{equation}  
ln(\frac{𝜇_i}{1-𝜇_i}) = -13.86x_{i1} -0.22x_{i2}  -2.30x_{i3}+ 2.49x_{i4}+ 1.01x_{i5}+ 1.05x_{i6}
\end{equation}  

The accuracy of this algorithm is 89.26% on the data from the test set, while the accurac on the training data is 92.95%. The execution time of the algorithm is 0 seconds.

###K-Nearest Neighbors method
In the space filled with data points whose position in space is determined by the value of their attributes. As it is necessary to determine the value of their dependent variable for each data point, the K closest points to each data point are observed. The instance is then classified to a class to which most of the data surrounding it belongs.

When applying this algorithm, the selection of the K parameter is extremely important. If this parameter is too small, overfitting of the model occurs. In this way, the model will overfit for train data set and will perform poorly with the test data. While too large parameter K will lead to underfitting the model because it looks at too much data leading to inaccurate results because the model cannot determine the rules by which to sort the data.


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(algorithm = 'brute', n_jobs=-1, n_neighbors=30)
knn.fit(X_train_new, y_train_new)
print("score on test: " + str(knn.score(X_test, y_test)))
print("score on train: "+ str(knn.score(X_train_new, y_train_new)))
print("of: "+ str(knn.score(X_train_new, y_train_new)-knn.score(X_test, y_test)))
k=np.r_[1:101]
knn_test_score=[]
knn_train_score=[]
knn_of=[]
print(k)
for i in k:
  knn_test_score.append(KNeighborsClassifier(algorithm = 'brute', n_jobs=-1, n_neighbors=i).fit(X_train_new, y_train_new).score(X_test, y_test))
print(knn_test_score)
for i in k:
  knn_train_score.append(KNeighborsClassifier(algorithm = 'brute', n_jobs=-1, n_neighbors=i).fit(X_train_new, y_train_new).score(X_train_new, y_train_new))
print(knn_train_score)

```

###Support Vector Machine method
Support Vector Machine, abbreviated SVM is an algorithm that finds a hyperplane in an N-dimensional space where N is the number of observed features that separate the data that is represented in the space as points based on their value of the parameter y. This hyperplane is an object in N-1 dimension that separates the points with the largest margin (distance between points from both classes). In a one-dimensional space, the hyperplane will be a point, in a two-dimensional space, a line, while in a three-dimensional space, it will be a two-dimensional plane.

Support vectors are points in space that are close to the hyperplane, and based on them, the plane is oriented and placed so that the margin is as large as possible. [17]

In this paper, the SVM algorithm is implemented using the LinearSVC model from the sklearn library, where the parameter C is set to 0.01, which represents the optimal value of the parameter C over this data set.

###Bagging Decision Tree Classifier method
Bagging is a term and a method coined by Leo Brayman from the phrase "bootstrap aggregating" and is used to improve the results of machine learning algorithms used for classification. This method achieves the reduction of variance (dispersion), which represents the sum of the squared differences of all values ​​from the set and their arithmetic mean. [23] In addition, it reduces the possibility of overfitting the model.

The data set is divided into random data groups on which a classification model is built. In this case, a decision tree is created for each group.
For each data item in the test data set, the value of the dependent variable is predicted based on all models. While the final value of the parameter y is assigned by determining the result given by the majority of the models. In this way, they are combined into one model that successfully classifies the data. [24,25,26]

In this paper, the Bagging Decision Tree Classifier algorithm was developed using the DecisionTreeClassifier and BaggingClassifier models from the sklearn library. The number of trees used in this model is 10. For this data set, the success of this algorithm is 99.36% on the data from the test set, while the success on the training data is 99.83%. The execution time of the algorithm is 0 seconds.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bg=BaggingClassifier(DecisionTreeClassifier(),max_samples=0.5,max_features=1.0,n_estimators=10)
bg.fit(X_train_new, y_train_new)
print("score on test: " + str(bg.score(X_test, y_test)))
print("score on train: "+ str(bg.score(X_train_new, y_train_new)))
```

###Ada Boosting Decision Tree Classifier method
Boosting is a modeling technique in machine learning that builds a powerful classifier using multiple weaker models by correcting the errors of the weaker model to make a better one. First, a model is built on the training data. Errors in this model are corrected and a new model is created. This process is repeated until all data from the test data set are predicted correctly or the maximum specified number of models is reached.

Since the original boosting algorithm did not adapt well enough and did not fully use the weak models, Shapir and Freund devised the AdaBoost algorithm. The name is an abbreviation of the name adaptive boosting algorithm. This was the first algorithm of its type to successfully perform binary classification and has been a very popular classification technique ever since.[25]

The AdaBoost classifier builds a classifier model based on an initial data set. Later, it applies a copy of the same classifier to the modified data set where the data that creates errors in the result is balanced so that the new classifier correctly classifies the data where it was previously incorrect. Data where the classifier made a mistake has its weight increased. More precisely, they behave as if there are more of them so that the algorithm does not ignore them. Since in this way, algorithm makes a mistake with new instances by applying a new model to the data, it is necessary to combine several different models. In this way, a new classifier is created which with high success classifies the data into two categories.[9]

In this paper, the AdaBoost algorithm was used, where the Decision Tree Classifier classification algorithm was used. This is implemented by the AdaBoostClassifier and DecisionTreeClassifier models from the sklearn library. The success of this algorithm is 99.57% on the data from the test set, while the success on the training data is 100%. The execution time of the algorithm is 0 seconds.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
adb = AdaBoostClassifier(DecisionTreeClassifier(min_samples_split=10,max_depth=4),n_estimators=10,learning_rate=0.6)
adb.fit(X_train_new, y_train_new)
print("score on test: " + str(adb.score(X_test, y_test)))
print("score on train: "+ str(adb.score(X_train_new, y_train_new)))
```

###Voting Classifier method
Voting Classifier is a classifier that groups data using already built models and combining their predictions. It determines the y value of an instance  by determining the probability of its belonging to a group by observing the results of previous models. As there are two ways of voting, soft and hard voting, there are two different ways in which the classifier determines the value of the parameter y based on the given models.

Hard Voting works on the principle of majority vote and the data is classified in the group in which the most models classified it
Soft Voting takes into account the probability calculated using each of the models to belong to a group and the result is obtained by calculating the average of all probabilities. [10]

In this paper, both voting methods gave the same percentages of success on training and testing data. The models used for Voting Classifier method are Decision Tree Classifier, Bagging Decision Tree Classifier and Ada Boosting Decision Tree Classifier because they gave the best results. The success of this algorithm is 99.85% on the data from the test set, while the success on the training data is 100%. The execution time of the algorithm is 0 seconds.

```python
from sklearn.ensemble import VotingClassifier
evc=VotingClassifier(estimators=[('adb',adb),('bg',bg),('clf',clf)],voting='hard')
evc.fit(X_train_new, y_train_new)
print("score on test: " + str(evc.score(X_test, y_test)))
print("score on train: "+ str(evc.score(X_train_new, y_train_new)))
```
